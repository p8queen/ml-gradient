```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['axes3d.mouserotationstyle'] = 'trackball'  # 'azel', 'trackball', 'sphere', or 'arcball'

def f(x,y):
    return 50 - (x**2 + y**2)

def grad_f(x,y):
    return np.array([-2*x, -2*y])
```

```python
n = 50
x_range = np.linspace(-10, 10, n)
y_range = np.linspace(-10, 10, n)
X, Y = np.meshgrid(x_range, y_range)
```

```python
plt.scatter(X, Y, c=f(X,Y), cmap='viridis')
plt.show()
```

```python
# fig size 
x_point = 9
y_point = -9
fig = plt.figure(figsize=(10, 10))
ax  = plt.axes(projection='3d')
#ax.plot_surface(X, Y, f(X,Y), cmap='viridis')
ax.plot_wireframe(X, Y, f(X,Y), color='black')
ax.scatter(x_point, y_point, f(x_point,y_point)+5, c='r', s=50)
for _ in range(10):
    grad = grad_f(x_point, y_point)
    x_point = x_point + 0.1 * grad[0]
    y_point = y_point + 0.1 * grad[1]
    ax.scatter(x_point, y_point, f(x_point,y_point)+5, c='b', s=50)
    
#ax.set_zlim(0, 70)
ax.view_init(10, 60)
plt.show()
```
