```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# funciones 
def f(x):
    return x**2

def df(x):
    return 2*x

def tangent_line(x, x0, y, slope):
    return slope*(x - x0) + y


```

```python
# punto inicial
x_ini = 6.5
x = np.linspace(-10,10,100) # crea un vector de 100 elementos entre -10 y 10
tangent = tangent_line(x, x_ini, f(x_ini), df(x_ini))

#graph

plt.title("Gradient Descent")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x,f(x), label='funcion')
plt.plot(x_ini,f(x_ini), 'ro', label='punto inicial', markersize=12)
plt.plot(x,tangent, 'g--', label='tangente')
plt.xlim(-10,10)
plt.ylim(-10,100)
plt.grid()
plt.legend()
plt.show()
```
