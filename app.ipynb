```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# funciones 
def f(x):
    return x**2

def df(x):
    return 2*x

def tangent_line(x, x0, y, slope):
    return slope*(x - x0) + y

def gradient_descent(x0, lr, epochs):
    # lr: learning rate, epochs: number of iterations, x0: initial guess
    x = x0
    x_list = [x]
    for _ in range(epochs):
        x = x - lr*df(x) # gradient is df(x), so the update is x - lr*df(x)
        x_list.append(x)
    return x_list
```

```python
# punto inicial
x_ini = 6.5
x = np.linspace(-10,10,100) # crea un vector de 100 elementos entre -10 y 10
tangent = tangent_line(x, x_ini, f(x_ini), df(x_ini))

# gradient descent
lr = 0.9
epochs = 30
x_list = gradient_descent(x_ini, lr, epochs)
#graph

plt.title("Gradient Descent")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x,f(x), label='funcion')
plt.plot(x_ini,f(x_ini), 'ro', label='punto inicial', markersize=12)
plt.plot(x,tangent, 'g--', label='tangente')
plt.plot(x_list, f(np.array(x_list)), 'bo-', label='iteraciones')
plt.plot(x_list[-1], f(x_list[-1]), 'go', label=f'{f(x_list[-1]):.4f}, last point')
plt.xlim(-10,10)
plt.ylim(-10,100)
plt.grid()
plt.legend()
plt.show()
```
